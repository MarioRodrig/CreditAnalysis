# 4. Linear Regression 

--------------

Linear regression is a fundamental technique employed for data analysis. It functions by modeling the relationship between a dependent variable and one or more independent variables. By fitting a linear equation to observed data, it allows for the assessment of how changes in the independent variables affect the dependent variable. This method involves estimating the coefficients of the linear equation to minimize the difference between the observed values and the values predicted by the model. Through this approach, linear regression can reveal insights into the strength and nature of relationships within the dataset, facilitating informed predictions and decisions.

# 5. Independent and dependent variable

Linear regression is a pivotal method in data analysis used to model the relationship between a dependent variable and one or more independent variables. In this approach, the dependent variable is the outcome or response that the model seeks to predict or explain, while the independent variables are the predictors or features that influence the dependent variable. By fitting a linear equation to the observed data, linear regression estimates the coefficients that quantify the relationship between the independent and dependent variables. This technique not only identifies the strength and direction of these relationships but also helps in making predictions and understanding how changes in the independent variables impact the dependent variable. The model's effectiveness is assessed through metrics such as R-squared and p-values, which indicate the goodness of fit and the statistical significance of the predictors.

# 8. Regression metrics for model performance

-------------

Regression metrics are crucial for evaluating the performance of a linear regression model. These metrics help determine how well the model fits the data and how accurately it predicts new outcomes. Here are some regression metrics ill be using. <br> 

#### 1.Mean Absolute Error <br>
- Measures the average magnitude of errors in a set of predictions, without considering their direction. It is the average of the absolute differences between predicted values and actual values. <br>

#### 2. Mean Squared Error <br>
- Measures the average squared difference between predicted values and actual values. It gives more weight to larger errors due to squaring.
  A lower MSE indicates better model performance. MSE is useful for identifying models with smaller prediction errors, but its units are the    square of the response variable’s units, which can be less intuitive. <br>

#### 3. Rooted Mean Squared Error <br>
- RMSE is the square root of MSE and provides a measure of the average magnitude of the errors in the same units as the response variable.
  A lower RMSE suggests a better fit. RMSE is often preferred over MSE as it is in the same units as the dependent variable, making it more     interpretable.

```python
from sklearn.metrics import mean_squared_error , mean_absolute_errobr
import math
```

```python
print("Absolute Mean Error: ", mean_absolute_error(y_test, predictions ))
print("Absolute squared Error: ", mean_squared_error(y_test, predictions ))
print("RMSE: ", math.sqrt(mean_squared_error( y_test, predictions )))
```

- High deviations in error indicate that the predictions are significantly inaccurate and unlikely to yield favorable outcomes.

# 9. Residual analysis

-----------

Residuals are the differences between the observed values and the values predicted by the linear regression model. They measure the error or deviation of the model's predictions from the actual data. By carefully analyzing residuals, you can identify potential problems with your model and make necessary adjustments to enhance its performance and reliability.

```python
residuals = y_test - predictions 
residuals
```

```python
sns.displot( residuals, bins= 15, kde= True )
plt.title("Residuals Analysis: Distribution and Density")
```

```python
import pylab
import scipy.stats as stats
stats.probplot (residuals, dist= "norm", plot= pylab)
pylab.show()
```

- Graph to Analyze Normality of Residuals
